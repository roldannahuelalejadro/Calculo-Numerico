```python
import numpy as np

def es_diag_dom(A):
    """
    Verifica si una matriz es estrictamente diagonal dominante
    """
    n = A.shape[0]
    for i in range(n):
        suma_fuera_diagonal = np.sum(np.abs(A[i])) - np.abs(A[i,i])
        if np.abs(A[i,i]) <= suma_fuera_diagonal:
            return False
    return True

def metodo_jacobi(A, b, x0, N):
    """
    Implementa el m√©todo de Jacobi
    """
    n = A.shape[0]
    
    # Verificar si A es diagonal dominante (condici√≥n suficiente para convergencia)
    if not es_diag_dom(A):
        print("Advertencia: La matriz no es estrictamente diagonal dominante.")
        print("La convergencia no est√° garantizada.")
    
    # Extraer componentes
    D = np.diag(np.diag(A))  # Matriz diagonal
    L_plus_U = A - D         # L + U
    
    # Matriz de iteraci√≥n de Jacobi
    D_inv = np.linalg.inv(D)
    T_j = -D_inv @ L_plus_U  # Matriz de iteraci√≥n
    c_j = D_inv @ b          # Vector constante
    
    # Verificar radio espectral (condici√≥n necesaria y suficiente)
    radio_espectral = np.max(np.abs(np.linalg.eigvals(T_j)))
    print(f"Radio espectral de T_j: {radio_espectral:.6f}")
    
    if radio_espectral >= 1:
        print("¬°ADVERTENCIA! Radio espectral >= 1. El m√©todo puede divergir.")
        return None
    
    # Iteraciones de Jacobi
    x = x0.copy()
    for i in range(N):
        x_new = T_j @ x + c_j
        # Calcular error relativo
        error = np.linalg.norm(x_new - x) / np.linalg.norm(x_new)
        if i % 10 == 0:  # Mostrar progreso cada 10 iteraciones
            print(f"Iteraci√≥n {i}: error = {error:.2e}")
        x = x_new
        
        # Criterio de parada opcional
        if error < 1e-10:
            print(f"Convergencia alcanzada en {i+1} iteraciones")
            break
    
    return x

# --- PRUEBA ---
# Tu matriz de ejemplo (CORREGIDA - la original ten√≠a filas repetidas)
matriz = np.array([ 
    [10, 2, 1],    # Cambiado para hacerla diagonal dominante
    [1, 8, 2], 
    [1, 1, 12] 
])

b = np.array([2, 3, 3])
x0 = np.array([1, 3, 0])

print("Matriz A:")
print(matriz)
print(f"\n¬øEs diagonal dominante? {es_diag_dom(matriz)}")

solucion = metodo_jacobi(matriz, b, x0, 50)

if solucion is not None:
    print(f"\nSoluci√≥n aproximada: {solucion}")
    print(f"Residuo: {np.linalg.norm(matriz @ solucion - b):.2e}")
    
    # Comparar con soluci√≥n exacta
    sol_exacta = np.linalg.solve(matriz, b)
    print(f"Soluci√≥n exacta: {sol_exacta}")
    print(f"Error: {np.linalg.norm(solucion - sol_exacta):.2e}")
```
