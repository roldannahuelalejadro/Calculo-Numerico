```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python

def metodo_biyeccion(y,a,b, N = 10):
    """ y(x) : funcion de x
        [a,b] espacio donde existe una raiz"""
    if y(a) * y(b) > 0:
        print("no cumple Bolzano")
        return None
    else:
        a_n = a
        b_n = b
        
        for i in range(N):
            x_n = (a_n+b_n)/2
            if y(x_n) == 0:
                print("Raiz encontrada en :",  x_n)
                return x_n
            elif y(x_n) < 0:
                a_n =  x_n
            elif y(x_n) > 0:
                b_n =  x_n
    return x_n
            
def y(x):
    t =  np.e**-x + x**3 + x**2 -5*x -4
    return t

r = metodo_biyeccion(y, 0, 5, 100)
print(r)
valores_x =  np.linspace(0,3  , 10000)
plt.plot(valores_x, y(valores_x))
plt.vlines(r, -1,1, "red")
plt.grid()
plt.show()

def y(x):
    t =  3 - (x+5)*np.e**-x
    return t

r = metodo_biyeccion(y, -4, -10, 100)
print(r)
valores_x =  np.linspace(-5, 3 , 10000)
plt.plot(valores_x, y(valores_x))
plt.vlines(r, -1,1, "red")
plt.grid()
plt.show()
```

```python
def newton_rapson(y,y_prima,x_0, N = 10):
    """ y: funcion de x, 
     [a,b] intervalo de x
       """
    x_n =  x_0
    for i in range(N):
        x_n =  x_n - y(x_n)/y_prima(x_n)
    return x_n

def y(x):
    t =  3 - (x+5)*np.e**-x
    return t

def y_prima(x):
    t =  (x+4)*np.e**-x
    return t

r = newton_rapson(y, y_prima, 5, 100)
print(r)
valores_x =  np.linspace(-5,10  , 100)
plt.plot(valores_x, y(valores_x))
plt.vlines(r, -1,1, "red")
plt.grid()
plt.show()

```
